Q5-> Create a NumPy array with random integers between 1 and 10 of shape (5,6) After creating the array
perform the following operations:

 a->Extract all even integers from array.

 b->Extract all odd integers from array

```python
x=np.random.randint(1,10,(5,6))
even=[]
odd=[]

for i in x:
  for j in i:
    if (j%2==0):
      even.append(j)
    else:
      odd.append(j)
print("Even",even)
print("Odd",odd)
```

Question->14

```python
dates = pd.date_range('2024-01-01', periods=100)
temperature = np.random.randint(20, 30, size=100)
humidity = np.random.randint(40, 60, size=100)

df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Temperature and Humidity Over Time')

fig.tight_layout()
plt.show()
```

Question->18

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Output to notebook (for Jupyter or Colab)
output_notebook()

# Define the data
x = range(0, 10)
y = [i for i in map(lambda x: round((x**2) * 0.1, 2), x)]  # Sine wave function

# Create a figure
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line plot
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3
p.grid.grid_line_color = 'navy'
p.grid.grid_line_dash = [6, 4]  # Dashed grid lines

# Show the plot
plt.show()
```
